```python
import requests
from bs4 import BeautifulSoup
import re
req = requests.get('https://news.daum.net/ranking/popular')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
titles = soup.select('a.link_txt')
info = soup.select('span.info_news')

news_titles = []
news_info = []

for i in titles:
    news_titles.append(i.text)
for dom in info:
    news_info.append(dom.text)


with open('news.csv', "wt", encoding="utf-8") as f:
    f.write('news_titles,news_info\n')  
    for i in range(50):
        f.write(news_titles[i]+","+news_info[i]+'\n')
        print(news_titles[i]+","+news_info[i]+'\n')
```
